```python
#Import the libraries for DBSCAN
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

# Import the Data

```python
Data=pd.read_csv('C:\\Users\\VivekG\\Desktop\\My Files- Vivek\\ExcelrThane\\DS\\Assignments\\Clustering\\crime_data.csv')
Data.head()
```

```python
Data.info()
```

```python
data1=Data.iloc[:,1:5]
```

```python
data1.values
```

```python
stscaler = StandardScaler().fit(data1.values)
X = stscaler.transform(data1.values)
```

```python
X
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=0.6,min_samples=4)
dbscan.fit(X)
```

```python
#Noisy samples Label = -1
dbscan.labels_
```

```python
cls=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cls
```

```python
    new_data=pd.concat([data1,cls],axis=1)
    new_data
```

```python
new_data.groupby('cluster').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(new_data['cluster'],new_data['Murder'], c=dbscan.labels_) 
```

```python
# Plot between pairs Murder~Assault
sns.lmplot( 'Murder','Assault', data=new_data,
        hue = 'cluster',
        fit_reg=False, size = 6 );
```

```python
# Plot between pairs Murder~Rape
sns.lmplot( 'Murder','Rape', data=new_data,
        hue = 'cluster',
        fit_reg=False, size = 6 );
```
