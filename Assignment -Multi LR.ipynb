# Model Validation

# Calculating VIF

```python
#calculate VIF=1/(1-Rsquare) for all independet variables

rsq_r = smf.ols('RNDS~ADM+MKTSP',data=data1).fit().rsquared  
vif_r = 1/(1-rsq_r)

rsq_a = smf.ols('ADM~RNDS+MKTSP',data=data1).fit().rsquared  
vif_a = 1/(1-rsq_a)

rsq_m= smf.ols('MKTSP~RNDS+ADM',data=data1).fit().rsquared  
vif_m = 1/(1-rsq_m) 


# Storing vif values in a data frame
d1 = {'Variables':['MKTSP','RNDS','ADM'],'VIF':[vif_m,vif_r,vif_a]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

none of the variable has VIF>20,No Collinearity,consider all variables in Regression equation.

# Residual Analysis

Test for Normality of Residuals (Q-Q Plot)

```python
import statsmodels.api as sm
qqplot=sm.qqplot(model.resid,line='q') # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-20000))     # Outlier detection as per above Q-Q plot
                
```

#Residual Plot for Homoscedasticity

```python
def get_standardized_values( vals ):
     return (vals - vals.mean())/vals.std()
    #Z=(x-mu)/sigma
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),
            get_standardized_values(model.resid))

plt.title('Residual Plot')
plt.xlabel('Standardized Fitted values')
plt.ylabel('Standardized residual values')
plt.show()
```

# Model Deletion Diagnostics

# Detecting Influencers/Outliers

Cooks Distance

```python
# Cook's Distance: If Cook's distance > 1, then it's an outlier
# influencers using cook's distance
model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
c
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(data1)), np.round(c, 7))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#index and value of influencer where c is more than 0.5
(np.argmax(c),np.max(c))
```

# High Influence points

```python
from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model)
plt.show()
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints
k = data1.shape[1]
n = data1.shape[0]
leverage_cutoff = 3*((k + 1)/n) #Points beyond Leverage_cutoff value are influencers
leverage_cutoff
```

From the above plot, it is evident that data point 49 is the influencers
